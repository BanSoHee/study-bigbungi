    - 올림/내림/버림 3가지 각각의 평균
        - 올림 : np.ceil()
        - 내림 : np.floor()
        - 버림 : np.trunc()
    - 모두 더하여 출력

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# data
df = pd.read_csv(r'C:\Project\study-bigbungi\데이터\Basic\basic1.csv')
df.head(3)
```

```python
# 올림/내림/버림 3가지 평균

# 올림
m_ceil = np.ceil(df['age']).mean()
# 내림
m_floor = np.floor(df['age']).mean()
# 버림
m_trunc = np.trunc(df['age']).mean()

print(f'올림 평균 : {m_ceil}')
print(f'내림 평균 : {m_floor}')
print(f'버림 평균 : {m_trunc}')
print(f'합계 : {m_ceil + m_floor + m_trunc}')
```
